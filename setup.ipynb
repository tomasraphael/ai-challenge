```python
import pandas
import sys
from os import listdir
import shutil
import os

listdir('./Lang/C++')
code_cnt = {}

for lang in listdir('./Lang/'):
    if lang == '.DS_Store':
        continue
    code_cnt[lang] = len(listdir('./Lang/' + lang))
    
#print(code_cnt)

df = pandas.DataFrame.from_dict(code_cnt, orient='index')
df = df.reset_index()
df.columns = ['Language Name', 'Count']
df = df.drop([203])
df = df.sort_values('Count', ascending=False)
df = df.head(30)
lang_array = df['Language Name'].values

files_df = pandas.DataFrame(columns=['Language Name', 'Path'], dtype=str)

i = 0
for lang in lang_array:
    if lang == '.DS_Store':
        continue
    for file in listdir('./Lang/' + lang):
        if file in ['.DS_Store', 'README', '00DESCRIPTION']: 
            continue
        files_df.loc[i] = [lang, './Lang/' + lang + '/' + file + '/' + listdir('./Lang/' + lang + '/' + file + '/')[0]]
        i+=1

files_df
    
```

```python
files_df = files_df.sample(frac=1., random_state=0)
files_df
```

```python
available_cnt = 0.8 * len(files_df)
available_cnt = int(available_cnt)
available_cnt

available_set = files_df.head(available_cnt)
non_available_set = files_df.tail(len(files_df) - available_cnt)
```

```python
if not os.path.exists('./challenge/dataset'):
    os.makedirs('./challenge/dataset')

if not os.path.exists('./challenge/evalset'):
    os.makedirs('./challenge/evalset')
    
if not os.path.exists('./challenge/answer'):
    os.makedirs('./challenge/answer')
    
for lang in lang_array:
    if not os.path.exists('./challenge/dataset/' + lang):
        os.makedirs('./challenge/dataset/' + lang)
```

```python
import shutil
available_set.apply(lambda r: shutil.copyfile(r['Path'], './challenge/dataset/' + r['Language Name'] + '/' + r['Path'].split('/')[-1]), axis=1)
```

```python
ans = []
for i, (idx, r) in enumerate(non_available_set.iterrows()):
    ans.append(r['Language Name'])
    shutil.copyfile(r['Path'], './challenge/evalset/' + str(i).zfill(4))
    
ans
```

```python
f = open('./challenge/answer/ans.txt', 'w')
f.write('\n'.join(ans))
f.close()
```
